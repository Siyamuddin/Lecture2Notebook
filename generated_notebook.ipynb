````
Here's the code based on the lecture transcript organized by topic:

### Topic 1: Full Loop
```python
# Print items in a list
fruits = ['apple', 'banana', 'cherry']
for F in fruits:
    print(F)

# Print numbers using the range function
for i in range(5):
    print(i)

# Print letters in a word
word = 'Python'
for L in word:
    print(L)
```

### Topic 2: While Loop
```python
# Loop until a condition is met
count = 0
while count < 5:
    print(count)
    count += 1

# Use break and continue in a while loop
n = 0
while n < 10:
    if n == 5:
        break
    else:
        print(n)
        n += 1

# Use continue in a while loop
n = 0
while n < 10:
    if n == 2:
        n += 1
        continue
    print(n)
    n += 1

# Loop with else
i = 0
while i < 3:
    print(i)
    i += 1
else:
    print('loop completed')

# Nested while loop
n = 0
while n < 3:
    i = 0
    while i < 3:
        if i == 2 and n == 2:
            print('i and j equals two, j')
        else:
            print(i)
        i += 1
    n += 1
```

### Topic 3: Comprehension with Loops
```python
# Create a list of square numbers
squares = [x**2 for x in range(4)]
for s in squares:
    print(s)
```

### Topic 4: Looping through Dictionary Items
```python
data = {'name': 'Alice', 'age': 25, 'city': 'New York'}
for key, value in data.items():
    print(f'{key}: {value}')
```

### Topic 5: Using Enumerate Function in Loops
```python
colors = ['red', 'blue', 'green']
for i, color in enumerate(colors):
    print(f'index: {i}, color: {color}')
```

### Topic 6: Using Zip Function
```python
colors = ['red', 'blue', 'green']
numbers = [1, 2, 3]
for color, number in zip(colors, numbers):
    print(f'color: {color}, number: {number}')
```

### Topic 7: Nested Loops in a List
```python
table = [[i*j for j in range(1, 11)] for i in range(1, 11)]
for row in table:
    print(row)
```

### Topic 8: Using Zip Function with Multiple Lists
```python
colors = ['red', 'blue', 'green']
numbers = [1, 2, 3]
types = ['shirt', 'pants', 'hats']
for color, number, type in zip(colors, numbers, types):
    print(f'color: {color}, number: {number}, type: {type}')
```

### Topic 9: Checking if All Numbers are Odd
```python
numbers = [1, 3, 5, 7, 9]
for number in numbers:
    if number % 2 == 0:
        print('found an even number')
        break
else:
    print('all numbers are odd')
```

### Topic 10: Infinite While Loop with Exit Condition
```python
while True:
    user_input = input('enter to exit, quit to quit: ')
    if user_input == 'quit':
        break
    else:
        print('you entered:', user_input)
```
````
